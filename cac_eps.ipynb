```python
import pandas as pd 
pd.set_option("display.max_rows",None)
pd.set_option('display.max_colwidth',1000) # para visualizar todo el texto de las columnas

def lower(text):
    return text.lower()

def normalize(s):
    replacements = (
        ("á", "a"),
        ("é", "e"),
        ("í", "i"),
        ("ó", "o"),
        ("ó", "o"),
        ("ú", "u"),
    )
    for a, b in replacements:
        s = s.replace(a, b)
    return s
        
def shorter_1(eps):
    eps = eps.replace('direccion departamental de','')
    eps = eps.replace('salud de','')
    eps = eps.replace('empresa medicina','')
    return eps

data = pd.read_csv('eps.csv')
data['eps'] = data['eps'].apply(lambda x: lower(x))
data['eps'] = data['eps'].apply(lambda x: normalize(x))
data['eps'] = data['eps'].apply(lambda x: shorter_1(x))
```
